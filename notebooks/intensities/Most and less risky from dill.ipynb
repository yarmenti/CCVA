#Computing mean risk for idiosync and groups

```python
idios = data['idio']
groups = data['groups']
```

```python
tmp = set()

for g in groups:
    print g["ids"]-tmp
    tmp = g["ids"]
    print
```

```python
means = []
for i in idios:
    id_ = list(i['id'])[0]    
    m = 0.
    hz = i['hazard_rates']
    m += hz[0] + hz[1]
    for g in groups:
        hz = g['hazard_rates']
        if id_ in g['ids']:
            m += hz[0] + hz[1]
        
    means.append({'id': list(i['id'])[0], 'mean': m*0.5})
```

```python
sorted_means = sorted(means, key=lambda k: k['mean']) 
sorted_means
```

```python
index_bank_to_choose = len(sorted_means)*2/3
print index_bank_to_choose
```
