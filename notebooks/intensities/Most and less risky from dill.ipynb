```python
import dill
```

```python
data = None
with open('intensities_no_I5.pkl', 'rb') as f:
    data = dill.load(f)
```

#Computing mean risk for groups

```python
groups = data['groups']

groups_means = []
for g in groups:
    hz = g['hazard_rates']
    mean = (hz[0]*3.+hz[1]*2.)/5.
    
    groups_means.append({'ids': g['ids'], 'mean': mean})
    
groups_means
```

#Computing mean risk for idiosync

```python
idios = data['idio']

idio_means = []
for i in idios:
    hz = i['hazard_rates']
    mean = (hz[0]*3.+hz[1]*2.)/5.
    
    idio_means.append({'id': list(i['id'])[0], 'mean': mean})

sorted(idio_means, key=lambda k: k['mean']) 
```

# Computing global risk

```python
glob_means = []

for i in idio_means:
    id_ = i['id']
    mean_idio = i['mean']
    
    mean_group = 0.
    for g in groups_means:
        if id_ in g['ids']:
            mean_group += g['mean']
            
    glob_means.append({'id': id_, 'mean': mean_idio+mean_idio})
```

```python
sorted_glob_mean = sorted(glob_means, key=lambda k: k['mean']) 
```
