# Model

We suppose that the default model of the each member is given by:

\begin{equation*}
\mathbb{P}(\tau_i \leq t) = 1 - \exp \left( - \sum_{i=0}^{k} \lambda_i (T_{i+1}-T_{i}) - \lambda_k (t-T_k) \right)
\end{equation*}

for $t \in [T_k, T_{k+1}[$, with the convention that $T_0 = 0$.

# Using Dataframe to get data

```python
import pandas as pd

df_intens = pd.DataFrame(calib_comm_shocks['Eta'])
print df_intens.columns
```

```python
df_intens = df_intens.drop([2, 3], axis=1)
df_intens.columns = ['3y_cum', '5y_cum']

indexes = df_intens.index

df_intens
```

```python
df_intens['mean_3y_5y_cum'] = 0.5*(df_intens['3y_cum'] + df_intens['5y_cum'])
df_intens
```

```python
all_mean_desc = df_intens.sort(columns=['mean_3y_5y_cum'], ascending=False)

indexes_all_mean_desc = all_mean_desc.index
print indexes_all_mean_desc
```

```python
plt.plot(indexes, all_mean_desc['mean_3y_5y_cum'].values, 'x', label='Mean 3y/5y desc ordering cum intensity')
plt.legend(loc='best')

plt.show()
```

# Get the groups structure

```python
groups = pd.DataFrame(calib_comm_shocks['Group_size']).transpose()
groups.columns = ['Nb_by_groups']
groups
```

```python
groups_intens = pd.DataFrame(calib_comm_shocks['Lambda_group_optimal'])
groups_intens.columns = ['3y_group', '5y_group']
groups_intens
```

```python
group_info = pd.concat([groups, groups_intens], axis=1)
group_info
```

```python
all_mean_desc['3y_group'] = 0
all_mean_desc['5y_group'] = 0

all_mean_desc.insert(0, 'ID', all_mean_desc.index)
all_mean_desc.index = np.arange(0, 125)

all_mean_desc
```

```python
for (nb, g_3y, g_5y) in zip(group_info['Nb_by_groups'], 
                            group_info['3y_group'], 
                            group_info['5y_group']):
    for i in range(0, nb):
        all_mean_desc.ix[i, '3y_group'] = all_mean_desc.ix[i, '3y_group'] + g_3y
        all_mean_desc.ix[i, '5y_group'] = all_mean_desc.ix[i, '5y_group'] + g_5y
```

```python
all_mean_desc
```

```python
all_mean_desc['3y_idio'] = all_mean_desc['3y_cum'] - all_mean_desc['3y_group']
all_mean_desc['5y_idio'] = all_mean_desc['5y_cum'] - all_mean_desc['5y_group']
```

```python
all_mean_desc = all_mean_desc.set_index('ID')
all_mean_desc
```

```python
plt.plot(indexes, all_mean_desc['3y_idio'].values, color='red', label='3y idiosync')
plt.plot(indexes, all_mean_desc['5y_idio'].values, color='green', label='5y idiosync')
plt.legend(loc='best')
plt.show()
```

```python
final_outputs = all_mean_desc[['3y_idio', '3y_group', '5y_idio', '5y_group']]
final_outputs
```
