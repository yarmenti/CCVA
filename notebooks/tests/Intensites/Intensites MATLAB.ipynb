# Model

We suppose that the default model of the each member is given by:

\begin{equation*}
\mathbb{P}(\tau_i \leq t) = 1 - \exp \left( - \sum_{i=0}^{k} \lambda_i (T_{i+1}-T_{i}) - \lambda_k (t-T_k) \right)
\end{equation*}

for $t \in [T_k, T_{k+1}[$, with the convention that $T_0 = 0$.

# Get the groups structure

```python
groups = pd.DataFrame(calib_comm_shocks['Group_size']).transpose()
groups.columns = ['Nb_by_groups']
groups
```

```python
groups_intens = pd.DataFrame(calib_comm_shocks['Lambda_group_optimal'])
groups_intens.columns = ['3y_group', '5y_group']
groups_intens
```

```python
group_info = pd.concat([groups, groups_intens], axis=1)
group_info
```

```python
cumul_3y_desc = pd.DataFrame(cumul_3y_desc)
cumul_3y_desc['3y_group'] = 0

cumul_3y_desc['ID'] = cumul_3y_desc.index
cumul_3y_desc.index = np.arange(0, 125)

cumul_3y_desc
```

```python
cumul_5y_desc = pd.DataFrame(cumul_5y_desc)
cumul_5y_desc['5y_group'] = 0

cumul_5y_desc['ID'] = cumul_5y_desc.index
cumul_5y_desc.index = np.arange(0, 125)

cumul_5y_desc
```

```python
cumul_3y_desc = pd.DataFrame(cumul_3y_desc)
cumul_3y_desc['3y_group'] = 0

cumul_5y_desc = pd.DataFrame(cumul_5y_desc)
cumul_5y_desc['5y_group'] = 0

for (nb, g_3y, g_5y) in zip(group_info['Nb_by_groups'], 
                            group_info['3y_group'], 
                            group_info['5y_group']):

    for i in range(0, nb):
        cumul_3y_desc.ix[i, '3y_group'] = cumul_3y_desc.ix[i, '3y_group'] + g_3y
        cumul_5y_desc.ix[i, '5y_group'] = cumul_5y_desc.ix[i, '5y_group'] + g_5y
```

```python
cumul_3y_desc['3y_idio'] = cumul_3y_desc['3y_cum'] - cumul_3y_desc['3y_group']
cumul_5y_desc['5y_idio'] = cumul_5y_desc['5y_cum'] - cumul_5y_desc['5y_group']
```

```python
cumul_3y_desc = cumul_3y_desc.set_index('ID')
cumul_3y_desc
```

```python
cumul_5y_desc = cumul_5y_desc.set_index('ID')
cumul_5y_desc
```

```python
global_intensities = pd.concat([cumul_3y_desc, cumul_5y_desc], axis=1)

# We check here that the mapping have been done 
# with IDS and not row by row

#0.126393 0.065811 0.060582
print cumul_3y_desc.ix[99, :]
print

#0.037683 0.021516 0.016167
print cumul_5y_desc.ix[99, :]
print
print global_intensities.ix[99, :]
```
