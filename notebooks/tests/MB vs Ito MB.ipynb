```python
import maths.montecarlo.path as path

from maths.montecarlo.process import (
    itoprocess, 
    brownianmotion
)
```

```python
h = 1/360.
time_grid = path.Path.generate_time_grid(0, 2, h)
    
vol = [0.1]
w0 = [0.]
```

```python
W = brownianmotion.StandardBrownianMotion(w0, vol, time_grid)
```

```python
import numpy as np

def drift_f(**kwargs):
    return np.array([0.])
        
def vol_f(**kwargs):
    return np.array([1.])
    
B = itoprocess.ItoProcessPath(w0, drift_f, vol_f, time_grid)
```

```python
%matplotlib inline
```

```python
W.plot()
```

```python
B.plot()
```

```python
%timeit -n 1000 W.simulate()
```
