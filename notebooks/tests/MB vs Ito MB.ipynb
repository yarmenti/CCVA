```python
import maths.montecarlo.path as path

from maths.montecarlo.processes import (
    itoprocess, 
    brownianmotion
)
```

```python
h = 1/360.
time_grid = path.Path.generate_time_grid(0, 2, h)
    
vol = 0.1    
w0 = 0.
```

```python
W = brownianmotion.StandardBrownianMotion(w0, vol, time_grid)
```

```python
def drift(**kwargs):
    return 0.
        
def vol(**kwargs):
    return 1.
    
B = itoprocess.ItoProcessPath(w0, drift, vol, time_grid)
```

```python
%matplotlib inline
```

```python
W.plot()
```

```python
B.plot()
```

```python
%timeit -n 10000 W.simulate()
```

```python
%timeit -n 10000 B.simulate()
```
