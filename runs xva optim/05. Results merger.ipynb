```python
import pandas as pd
import numpy as np
import os

final_root_path = './res/aggregated/'

res_root_path = './res/sim%i'

loss_path = 'loss_1y_ahead'
im_path = 'im'
cva_path = 'cva_ccp'
```

```python
import itertools
```

```python
loss_ahead_dict = dict()
im_dict = dict()
cva_dict = dict()

for i in itertools.count():
    if i == 0:
        continue
    
    test_path = res_root_path % i
    
    if os.path.isdir(test_path):
        crt_loss_path = os.path.join(test_path, loss_path)
        for fname in os.listdir(crt_loss_path):
            if fname not in loss_ahead_dict:
                loss_ahead_dict[fname] = pd.DataFrame()
            
            path_file = os.path.join(crt_loss_path, fname)
            print path_file
            pd_df_ = pd.read_csv(path_file, index_col=0, header=0)
            loss_ahead_dict[fname] = loss_ahead_dict[fname].append(pd_df_, ignore_index=True)
            
        crt_im_path = os.path.join(test_path, im_path)
        for dname in os.listdir(crt_im_path):
            if dname not in im_dict:
                im_dict[dname] = dict()
                
            crt_im_path_inside = os.path.join(crt_im_path, dname)
            for fname in os.listdir(crt_im_path_inside):
                if fname not in im_dict[dname]:
                    #im_dict[dname][fname] = []
                    im_dict[dname][fname] = pd.DataFrame()
            
                path_file = os.path.join(crt_im_path_inside, fname)
                print path_file
                #im_dict[dname][fname].append(path_file)
                pd_df_ = pd.read_csv(path_file, index_col=0, header=0)
                im_dict[dname][fname] = im_dict[dname][fname].append(pd_df_, ignore_index=True)
                
        crt_cva_path = os.path.join(test_path, cva_path)
        for dname in os.listdir(crt_cva_path):
            if dname not in cva_dict:
                cva_dict[dname] = dict()
                
            crt_cva_path_inside = os.path.join(crt_cva_path, dname)
            for fname in os.listdir(crt_cva_path_inside):
                if fname not in cva_dict[dname]:
                    #im_dict[dname][fname] = []
                    cva_dict[dname][fname] = pd.DataFrame()
            
                path_file = os.path.join(crt_cva_path_inside, fname)
                print path_file
                #im_dict[dname][fname].append(path_file)
                pd_df_ = pd.read_csv(path_file, index_col=0, header=0)
                cva_dict[dname][fname] = cva_dict[dname][fname].append(pd_df_, ignore_index=True)        
    else:
        break
```

```python
for k, v in loss_ahead_dict.iteritems():
    path = os.path.join(final_root_path, loss_path + '/', k)
    print path
    v.to_csv(path)
    
print
    
for k, v in im_dict.iteritems():
    path = os.path.join(final_root_path, im_path + '/', k)
    print path
    
    if not os.path.isdir(path):
        os.makedirs(path)
    
    for kk, vv in v.iteritems():
        path_f = os.path.join(path, kk)
        print path_f
        vv.to_csv(path_f)
        
print

for k, v in cva_dict.iteritems():
    path = os.path.join(final_root_path, cva_path + '/', k)
    print path
    
    if not os.path.isdir(path):
        os.makedirs(path)
    
    for kk, vv in v.iteritems():
        path_f = os.path.join(path, kk)
        print path_f
        vv.to_csv(path_f)
```

```python
pd_df = cva_dict.values()[0].values()[0].copy()
pd_df[:] = 0.

path = os.path.join(final_root_path, cva_path + '/0.85/', "cva_5.00.csv")
pd_df.to_csv(path)
```

```python
default_times_DF = pd.DataFrame()

for i in itertools.count():
    if i == 0:
        continue
    
    test_path = res_root_path % i
    
    if os.path.isdir(test_path):
        tmp_path = os.path.join(test_path, 'df_times.csv')
        tmp_df_times = pd.read_csv(tmp_path, index_col=0, header=0).transpose()
        default_times_DF = default_times_DF.append(tmp_df_times, ignore_index=True)
    else:
        break
        
default_times_DF = default_times_DF.transpose()

default_times_DF.to_csv(final_root_path + 'default_times.csv')
```
